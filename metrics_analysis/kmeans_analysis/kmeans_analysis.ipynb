```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('john6_bench_cluster.csv') 
```

```python
df =data

df["ratio"] = df[["y_cuml","y_sklearn"]].apply(lambda x:x[1]/x[0], axis=1)
```

```python
styles = ["-","-."]
df[["y_cuml","y_sklearn"]].plot(logy=True,style=styles)
plt.xlabel("Nombre d'instances (*50k)")
plt.ylabel("Temps en s")
plt.grid(True, which="major", ls="-")
plt.grid(True, which="minor", ls="--")
# ajouter john1 /john6
plt.savefig('test_kmeans_john6.svg') 
```

```python
df["ratio"].plot()
plt.xlabel("Nombre d'instances (*50k)")
plt.ylabel("Ratio tps_sklearn sur tps_cuml")
plt.grid(True, which="major", ls="-")
# start avec y=0
plt.ylim(bottom=0)
plt.savefig('test_kmeans_ratio_john6.svg') 
```
