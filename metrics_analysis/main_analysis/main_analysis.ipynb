```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('metrics_rapids_18_5_100_50.csv') 
#data = pd.read_csv('metrics_25_5000_500.csv') 
#data = pd.read_csv('metrics_25_2500_500.csv') 
data['timestamp'] = pd.to_datetime(data['timestamp'])
df = data
data = data.resample("1S", on="timestamp").sum()
data = data[(data.T != 0).any()]
data["time_total"] = data[["time_load","time_train"]].sum(axis=1)
data["accu_train"] = data["size_batch_training"].cumsum()
data["accu_sent"] = data["nb_url_sent"].cumsum()

```

```python
data[["time_load","time_train"]].plot()
```

```python
data[["size_batch_training","nb_url_sent"]].plot()
```

```python
data[["accu_train","accu_sent"]].plot()
```
