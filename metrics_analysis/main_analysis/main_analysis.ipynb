```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('5_john6_metrics_rapids_3_5_100_50.csv') 
#data = pd.read_csv('metrics_25_5000_500.csv') 
#data = pd.read_csv('metrics_25_2500_500.csv') 
data['timestamp'] = pd.to_datetime(data['timestamp'])
df = data
data = data.resample("1S", on="timestamp").sum()
data['tvalue'] = data.index
data['delta'] = (data['tvalue']-data['tvalue'][0])
data['delta'] =data['delta'].apply(lambda x:int( x.value/ 10**9))
data.set_index('delta', inplace=True)
data = data[(data.T != 0).any()]
data = data[(data.T != 0).any()]
data["time_total"] = data[["time_load","time_train"]].sum(axis=1)
data["accu_train"] = data["size_batch_training"].cumsum()
data["accu_sent"] = data["nb_url_sent"].cumsum()

```

```python
df=data[["size_batch_training","nb_url_sent"]]


styles = ['-.',"-"]

df.plot(style=styles)
plt.xlabel("Temps en seconde")
plt.ylabel("Nombre d'urls/s")
plt.grid(True, which="major", ls="-")
plt.ylim(bottom=0)
# start avec y=0
plt.savefig('test_john6_3.svg') 
```
